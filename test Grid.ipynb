To get a good approximation of 4th derivatives, the number of grid points must be large enough. This is because we use polynomials to approximate, whose order reduces after differentiation. The numerical experiments below show that there should be at least 8 grid points in each directon.

```python
import numpy as np
from numpy import pi, cos, sin, log, sqrt, exp
import matplotlib.pyplot as plt
from matplotlib import cm

# Reloading the module
import importlib

import SEM2D_Grid
importlib.reload(SEM2D_Grid)
from SEM2D_Grid import *
```

```python
grid = Grid(Nx=8, Ny=8, node_type_x="Legendre", node_type_y="Chebyshev")
```

```python
u = exp(grid.nodes_x)
u_deri_th = exp(grid.nodes_x)
u_deri = np.dot(grid.Dx_list[4], u)
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(grid.nodes_x, u_deri)
ax.scatter(grid.nodes_x, u_deri_th)
```

```python
u = sin(grid.nodes_y)
u_deri_th = sin(grid.nodes_y)
u_deri = np.dot(grid.Dy_list[4], u)
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(grid.nodes_y, u_deri)
ax.scatter(grid.nodes_y, u_deri_th)
```
